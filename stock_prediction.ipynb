```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('new_train.csv')
data.head()
```

```python
x=pd.DataFrame(data, columns=['Open-Stock-1','High-Stock-1','Low-Stock-1','VWAP-Stock-1','Volume-Stock-1','Turnover-Stock-1'])
y=pd.DataFrame(data, columns=['Close-Stock-1'])
x.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(x);
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scalar= StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
x_train=scalar.fit_transform(x_train)
x_test=scalar.fit_transform(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
```

```python
reg = GradientBoostingRegressor(random_state=1)
reg.fit(x_train,y_train.values.ravel())
```

```python
y_pred= np.round(reg.predict(x_test))
y_pred
```

```python
import numpy as np
np.round(reg.score(x_test, y_test.values.ravel()))
```

```python
np.round(mean_squared_error(y_test, y_pred))
```

```python
np.round(mean_squared_error(y_test, y_pred, squared=False))
```
